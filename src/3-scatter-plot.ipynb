### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt

# Define x and corrected y values
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]  # Corrected values

# Error handling
if len(x) != len(y):
    raise ValueError("X and Y must have the same length.")

# Create scatter plot
plt.scatter(x, y, color='red', label="y = 2x")

# Customize the plot
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
